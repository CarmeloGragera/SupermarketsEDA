```python
import pandas as pd
```

```python
cerveza_C = pd.read_csv('c_carr')
cerveza_D = pd.read_csv('c_dia')
cerveza_M = pd.read_csv('c_merc')
```

```python
cerveza_C = cerveza_C.drop(['Unnamed: 0'],axis=1)
```

```python
cerveza_C = cerveza_C.rename(columns={'price':'Carrefour'})
```

```python
cerveza_D = cerveza_D.drop(['Unnamed: 0'],axis=1)
```

```python
cerveza_M = cerveza_M.drop(['Unnamed: 0'],axis=1)
```

```python
cerveza = pd.merge(cerveza_M,cerveza_C, how='outer')
cerveza = pd.merge(cerveza, cerveza_D, how='outer')
```

```python
tomate_C = pd.read_csv('carrefour_tomate.csv')
tomate_D = pd.read_csv('dia_tomate.csv')
tomate_M = pd.read_csv('mercadona_tomate.csv')
```

```python
tomate = pd.merge(tomate_C,tomate_D, how='outer')
tomate = pd.merge(tomate,tomate_M, how='outer')
```

```python
total = pd.concat([tomate] + [cerveza])
```

```python
dia = pd.read_csv('DIA_huevos_cafe.csv')
merc = pd.read_csv('Mercadona_huevos_cafe.csv')
carr = pd.read_csv('carrefour_huevos_cafe.csv')
```

```python
dia_HC = dia.drop(['Unnamed: 0'],axis=1)
dia_HC = dia_HC.drop(['reference_unit'],axis=1)
dia_HC = dia_HC.drop(['category'],axis=1)
dia_HC = dia_HC.rename(columns={'price':'DIA'})
```

```python
merc_HC = merc.drop(['Unnamed: 0'],axis=1)
merc_HC = merc_HC.drop(['reference_unit'],axis=1)
merc_HC = merc_HC.drop(['category'],axis=1)
merc_HC = merc_HC.rename(columns={'price':'Mercadona'})
```

```python
carr_HC = carr.drop(['Unnamed: 0'],axis=1)
carr_HC = carr_HC.drop(['reference_unit'],axis=1)
carr_HC = carr_HC.drop(['category'],axis=1)
carr_HC = carr_HC.rename(columns={'price':'Carrefour'})
```

```python
cerv_huevos = pd.merge(dia_HC,merc_HC, how='outer')
cerv_huevos = pd.merge(cerv_huevos,carr_HC, how='outer')
```

```python
total = pd.concat([total] + [cerv_huevos])
```
